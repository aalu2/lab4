```python
#Answer of 1
def funny(n):
  for i in range(1,n+1):
    if i % 4 == 0 and i % 6 == 0:
      print("funny buzz")
    elif i % 4 == 0:
      print("funny")
    elif i % 6 == 0:
      print("buzz")
    else:
      print(i)
funny(16)  
```

```python
#Answer of 2
def mul_sum(m,n,limit):
  sum=0
  for i in range(0,limit):
    if(i%m==0) or (i%n==0):
      sum=sum+i
  print(sum)
mul_sum(3,5,10)
mul_sum(3,7,1000)
```

```python
#Answer of 3
def hailstone(m):
  while(m!=1):
    if m%2==0:
      m=int(m/2)
      print(m)
    else:
     m=m*3+1
     print(m)
hailstone(11)
```

```python
#Answer of 4
def nearest(n):
  i=0
  lst=[]
  if(n>1):
    while True:
      res=2**i
      i=i+1
      if res>n:
        lst.append(res)
        break
    lst.append(2**(i-2))
    lst.sort()
  elif(n<1):
    while True:
      res=2**i
      i=i-1
      if res<n:
        lst.append(res)
        break
    lst.append(2**(i+2))
    lst.sort()
  print(lst)
   
  if (lst[1]-n<n-lst[0]):
      print(lst[1])
  else:
      print(lst[0])
nearest(0.1) 
nearest(11.5)
nearest(2016)


```

```python
#Answer of 5
def check(m):
  lst=[]
  if(m>0):
    for i in range(m,0,-1):
     if(m % i == 0):
      lst.append(i)
    lst.sort()
    print(lst)
    for i in range(1,int(len(lst)/2)+1):
        print(lst[i-1],"*",lst[-i])
        lst.remove(m)
        print(lst)
        if sum(lst)>m:
          return True
        else:
          return False
print(check(12))
print(check(14))
print(check(16))
print(check(20))
print(check(22))
print(check(24))
```
