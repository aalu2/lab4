```python
# Answer of 3
def hailstone(n):
    
    print(n)
    if n == 1:
        return 1
    elif n % 2 == 0:
        return 1 + hailstone(n // 2)
    else:
        return 1 + hailstone(1+n * 3)
print(hailstone(17))
print(hailstone(20))
print(hailstone(8))
```

```python
#aNSWER OF 4
#Definition of the nearest() function which takes one integer called n.
def nearest(n):
    
    #Setting the value of i to 1...
    i = 1.0
    
    #Checking if the n is greater than one...
    if n > 1:
        
        #Updating the value of i...
        while (2**i) < n:
            i+=1
            
        #Finding the distance between two values from n...
        dist1 = n - 2**(i-1)
        dist2 = 2**i - n
        
        #As per the values of dist1 and dist2, return the answer...
        if dist1 > dist2:
            return 2**(i)
        elif dist1 == dist2:
            return max(2**i, 2**(i-1))
        else:
            return 2**(i-1)
            
    #Otherwise, go for 1/2^i...
    else:
        
        #Setting the value of i to 0...
        i = 0
        
        #Updating the value of i...
        while True:
            temp = 1/(2**i)
            if temp < n:
                break
            else:
                i+=1
        
        #Finding the dist1 and dist2 from n...
        dist1 = 1/(2**i) - n
        dist2 = n - 1/(2**(i+1))
        
        #As per the values of dist1 and dist2, return the answer...
        if dist1 > dist2:
            return 1/(2**(i))
        elif dist1 == dist2:
            return max(1/(2**(i)), 1/(2**(i-1)))
        else:
            return 1/(2**(i-1))
        

#Call the function and print the results...
print('nearest(8) =', nearest(8))
print('nearest(11.5) =', nearest(11.5))
print('nearest(14) =', nearest(14))
print('nearest(2016) =', nearest(2016))
print('nearest(0.1) =', nearest(0.1))
print('nearest(0.75) =', nearest(0.75))
print('nearest(1.5) =', nearest(1.5))
```
